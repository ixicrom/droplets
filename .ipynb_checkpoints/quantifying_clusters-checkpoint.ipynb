# Quantifying clusters
Aims:
* Find a numerical measure of how pure each original image is, in terms of which clusters it is sorted into
* The ideal case is that each image is (almost) entirely in one cluster
* Avoid biasing the system towards anything other than this goal
* Hopefully use this measure to determine how many clusters is optimal for k-means clustering

## Function to calculate Gini index measure of node purity

```python
def gini_original(cluster_freq):
    sample_scores = list() # make list for output scores
    
    for col in cluster_freq.columns: # will calculate a score for each image
        counts = cluster_freq[col].values
        p = counts[counts!=0]/12 # remove all zeroes (those clusters aren't present to don't want to include) and normalise
        q = 1-p
        G = sum(p*q) # calculate index
        sample_scores.append(G) 
    return sample_scores
```

Taking inspiration from its use in decision trees, I calculated this measure viewing each original image as a equivalent to a decision tree node. This raises some issues:
### Should I use the Gini index or the entropy?
The book says they are numerically basically the same, and this seems to be the case with my data too. I'll use the Gini index because it stays below 1 and is therefore a bit more intuitive to me.
### How should the score for each image be combined to make an overall score? 
I chose to take the mean as taking a sum makes sense as a simple combination, and the mean is equivalent but has the added benefit that the scale of the measure is preserved.
### How does the measure deal with a changing number of clusters?
This is where things start to become tricky. With more clusters possible, there is an increased chance of more clusters being present in the image. This may cause issues. I can think of some ways around this though:
* Divide by the number of clusters (or some power of it?) to somewhat reduce the effect


```python
def gini_normalised(cluster_freq, power):
    sample_scores = list()
    n_clust = cluster_freq.shape[0]
    for col in cluster_freq.columns:
        counts = cluster_freq[col].values
        p = counts[counts!=0]/12
        q = 1-p
        G = sum(p*q)
        sample_scores.append(G/n_clust**power)
    return sample_scores
```

* Only include the largest n clusters in the score, so the later entries aren't affected so adversely


```python
def gini_truncated(cluster_freq):
    sample_scores = list()
    
    for col in cluster_freq.columns:
        counts = cluster_freq[col].values
        counts
        p = counts[counts!=0]/12
        p_sorted = np.sort(p)
        p_short = p_sorted[:3]
        p_short
        q_short = 1-p_short
        G = sum(p_short*q_short)
        sample_scores.append(G)
    return sample_scores
```

* Decide that I shouldn't change it, it is probably designed this way for a purpose so maybe I'm just abusing it and it isn't appropriate for comparing different cluster sizes. Would still be a useful tool for comparing different models

```python
scores_orig = list()
scores_norm = list()
scores_trunc = list()
for n_clust in range(2,100):
    np.random.seed(1234)
    kmeans=cluster.KMeans(n_clusters=n_clust).fit(dat_forLearning)
    labs=kmeans.labels_
    k_clusters = pd.Series(labs, index = dat_forLearning.index, name='Cluster_kmeans')

    dat_results = pd.concat([dat_forLearning, k_clusters], axis=1)
    dat_results = dat_results.set_index('Cluster_kmeans', append = True)

    dat_toPlot = dat_results.stack().reset_index()
    dat_toPlot.columns = ['Slice', 'Colour', 'Cluster_kmeans', 'r', 'Value']
    dat_toPlot['Sample'] = dat_toPlot['Slice'].str.slice(0,8)
    dat_toPlot = dat_toPlot.drop(['r', 'Value'], axis='columns').drop_duplicates().reset_index(drop=True)
    cl_count = count_clusters(dat_toPlot, grouper1='Sample', grouper2='Colour', counter='Cluster_kmeans')
    # cl_count.transpose().plot(kind='bar', stacked = True, title=str(n_clust), legend=False)

    score_orig = np.mean(gini_original(cl_count))
    scores_orig.append(score_orig)

    score_norm = np.mean(gini_normalised(cl_count, 0.5))
    scores_norm.append(score_norm)

    score_trunc = np.mean(gini_truncated(cl_count))
    scores_trunc.append(score_trunc)
```
